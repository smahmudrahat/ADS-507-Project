```python
# Re-load the data
data = df_title

# Remove the 'endYear' column and replace '\N' with NaN
data.drop(columns=['endYear'], inplace=True)
data.replace('\\N', pd.NA, inplace=True)

# Split the 'genres' column into separate rows
# First, we convert the 'genres' string into a list of genres
data['genres'] = data['genres'].str.split(',')

# Now, we 'explode' the genres column
df_title = data.explode('genres')

# Display the first few rows of the modified dataset
df_title.head()
```

```python
data_names = df_name

# Step 1: Replace '\N' with NaN
data_names.replace('\\N', pd.NA, inplace=True)

# Step 2: Split the 'primaryProfession' column into separate rows
data_names['primaryProfession'] = data_names['primaryProfession'].str.split(',')

# Step 3: Split the 'knownForTitles' column into separate rows
data_names['knownForTitles'] = data_names['knownForTitles'].str.split(',')

# Step 4: Explode the 'primaryProfession' column
data_names = data_names.explode('primaryProfession')

# Step 5: Explode the 'knownForTitles' column
data_names = data_names.explode('knownForTitles')

# Replace the datanames with our prevous dataFrame
df_name = data_names

# Display the first few rows of the modified dataset
df_name.head()
```

### Now we will work with df_aska

```python
# Re-load the dataframe
df_aska = pd.read_csv(path_akas, nrows=50000)

data_names = df_aska

# Step 1: Replace '\N' with NaN
data_names.replace('\\N', pd.NA, inplace=True)

# number of missing value in attributes value from data_names
missing_value = data_names['attributes'].isna().sum()

# percentage of missing value
percentage_missing_value = (missing_value/data_names.shape[0])*100

print(f"Number of missing value {missing_value}, from total {data_names.shape[0]}, percentage of missing value {percentage_missing_value}%")

# number of missing value in language value from data_names
missing_value = data_names['language'].isna().sum()

# percentage of missing value
percentage_missing_value = (missing_value/data_names.shape[0])*100

print(f"Number of missing value {missing_value}, from total {data_names.shape[0]}, percentage of missing value {percentage_missing_value}%")

# delete attribute form the dataframe
data_names.drop(columns=['attributes', 'language'], inplace=True)

# Replace the datanames with our prevous dataFrame
df_aska = data_names

# Display the first few rows of the modified dataset
df_aska.head()
```
